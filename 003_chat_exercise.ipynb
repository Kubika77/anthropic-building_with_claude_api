### Create a chat boc with memory

```python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()


def add_user_message(messages: list[dict], msg: str) -> None:
    """
    Add a user message to the messages list.
    
    Args:
        messages (list[dict]): The list of messages.
        msg (str): The user message to add.
    """
    messages.append({"role": "user", "content": msg})

def add_assistant_message(messages: list[dict], msg: str) -> None:
    """
    Add an assistant message to the messages list.
    
    Args:
        messages (list[dict]): The list of messages.
        msg (str): The assistant message to add.
    """
    messages.append({"role": "assistant", "content": msg})

def chat(messages: list[dict]) -> str:
    """
    Send a list of messages to the API and return the assistant's response.
    
    Args:
        messages (list[dict]): The list of messages to send.
    
    Returns:
        str: The assistant's response.
    """
    anthropic = Anthropic()
    response = anthropic.messages.create(
        model="claude-haiku-4-5",
        max_tokens=100,
        messages=messages
    )
    return response.content[0].text

messages = []
while True:
    user_input = input("Enter your question (or 'exit' to quit): ")
    print(f"User input: {user_input}")
    if user_input.lower() == "exit":
        break
    add_user_message(messages, user_input)
    assistant_response = chat(messages)
    print(f"Assistant response: {assistant_response}")
    add_assistant_message(messages, assistant_response)

print(f"Messages: {messages}")

```

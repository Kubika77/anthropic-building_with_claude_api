### System prompts
System prompts provide the llm guidance on how to response.
LLM will try to respond in the same way someone in the specified role would response.
Help keep LLM on track

```python
from anthropic import Anthropic
from dotenv import load_dotenv

load_dotenv()


def add_user_message(messages:list[dict], msg:str) -> None:
    messages.append({"role": "user", "content": msg})

def add_assistant_message(messages:list[dict], msg:str) -> None:
    messages.append({"role": "assistant", "content": msg})

def chat(messages:list[dict], system_prompt:str | None = None) -> str:
    params = {
        "model": "claude-haiku-4-5",
        "messages": messages,
        "max_tokens": 1000
    }
    if system_prompt:
        params["system"] = system_prompt

    client = Anthropic()
    response = client.messages.create(**params)

    return response.content[0].text

system_prompt = """
You are a patient Math tutor. Do not directly answer a student question.
Guide them to a solution step by step.
"""

messages = []

user_message = "How do I solve 5x+3=2 for x?"
add_user_message(messages, user_message)
print(f"user question: {user_message}")
answer = chat(messages, system_prompt)
print(f"Math tutor guidance is: {answer}")
```

#### System prompt exercise

```python
messages = []
user_message = """
Write a python function that checks a string for duplicate characters
"""

system_prompt = """
You are a senior python developer who writes very concise code.
give very short and concise explanation.
"""

add_user_message(messages, user_message)
print(f"user question: {user_message}")
chat_response = chat(messages, system_prompt)
print(f"Python developer guidance is: {chat_response}")
```

#### Multi turn conversations:
remember the previous messages for a continous conversation

##### example when messages are not related (no memory of any previous messages)

```python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
client = Anthropic()

message = client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=100,
    messages=[
        {"role": "user",
         "content": "What is quantom computing? answer in one sentence"}
    ]
)

print(f"first answer is: {message.content[0].text}")

message_2 = client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=100,
    messages=[
        {"role": "user",
         "content": "write another sentence."}
    ]
)
print(f"second answer is: {message_2.content[0].text}")
```

##### Example with multi turn conversation (memory of previous messages):
Saving the entire messages history (user and AI messages) inside a list of messages, and providing it into the client messages as input.

```python
def add_user_message(messages: list[dict], msg: str) -> None:
    user_msg = {
        "role": "user",
        "content": msg
    }
    messages.append(user_msg)

def add_assistant_message(messages: list[dict], msg: str) -> None:
    assistant_msg = {
        "role": "assistant",
        "content": msg
    }
    messages.append(assistant_msg)

def chat(messages: list[dict]) -> str:
    response = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=100,
        messages=messages
    )
    return response.content[0].text
```

```python
# Make a starting list of messages:
messages = []

# Add in the initial user message
add_user_message(messages, "What is quantum computing? answer in one sentence")

print(f"messages: {messages}")

# pass the list of messages into 'chat' to get an answer
answer = chat(messages)
print(f"answer: {answer}")
```

```python
# Take the answer and add it to the messages list as an assistant message
add_assistant_message(messages, answer)

# Add in user follow-up message and get another answer
add_user_message(messages, "write another sentence.")

# Call the 'chat' function again with the updated messages list
answer = chat(messages)
print(f"answer: {answer}")
```
